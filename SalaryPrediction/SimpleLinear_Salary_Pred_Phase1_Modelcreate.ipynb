```python
#Simple Linear Algorithm - Model Creation - Phase 1
#Salary_prediction used to predict the salary for all experience
#pandas is the library which has all functions, classes

import pandas as pd
```

```python
# pd is having all library and pd is calling the read_csv funtion to read the salary.csv file and stores into dataset variabe
dataset = pd.read_csv("salary.csv")
```

```python
# returns the values loaded into dataset variable
dataset
# Data collection completes till here
```

```python
# Input and Output Spits starts

# Defining and calling the input values - independent is the input
independent = dataset [["YearOfExperience"]]
```

```python
# Defining and calling the Output values - dependent is the output or expected Values
dependent = dataset [["Salary"]]

# Input and Output Spits ends
```

```python
# Split train and test
# import model_selection class from sklearn library and import train_test_split function inside model_selection class
# sklearn provider is provinding the class

from sklearn.model_selection import train_test_split

# passing the independent, dependent, test data size and randomstate into train_test_split
# assigning the independent, dependent into X_train,x_test, y_train, y_test - means allocating input for both training and testing purpose and the same applies for the output.
# below code displays in the array_lists with comma for train and test sets.

X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size = 0.30, random_state=0)

# Input and Output Spits Ends
```

```python
#Model Creation Starts:
#import linear_model class from sklearn
# Functions available inside LinearRegression class will be assigned to regressor
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
LinearRegression()
```

```python
# Standard syntax to get Weight
# coef_ is holding the formula to calculate the weight
weight = regressor.coef_
weight
```

```python
# standard syntax to get Bias
# intercept_ is holding the formula to calculate the bias
bias = regressor.intercept_
bias

#Model Creation Ends:
```

```python
#Evaluation Metrics starts
# predict is the function for the prediction and it has to be done for X_test value
#Evaluation Value will be stored in y_pred
y_pred = regressor.predict(X_test)
```

```python
# metrics is the class available in sklearn and import r2_score
# To calculate r2_score, both test input and output passed as an input 
# r_score is the variable assigned
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
r_score
# r_score is near to 1 which represents the good Model
# Eveluation Ends
```

```python
# Save the Good Model
#file saved in the same location
# wb - write to binary
import pickle
filename ="finalized_model_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```
